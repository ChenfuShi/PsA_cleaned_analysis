```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['svg.fonttype'] = 'none'
base_dir = "http://bartzabel.ls.manchester.ac.uk/orozcolab/SNP2Mechanism/"
```

```python
df = pd.read_csv("qc_results.txt", sep = ',')
```

```python
metadata_ATAC = pd.read_csv(f"{base_dir}/metadata/cleaned_ATAC_metadata.csv", index_col=0)
```

```python
df['sample'] = df['sample'].str.replace('_ATAC','')
```

```python
df = df.merge(metadata_ATAC, left_on = 'sample', right_on = 'id')
```

```python
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(18, 10))  # Adjust the size as needed
barplot = sns.barplot(data=df, x="proper_name", y="reads_in_tss", palette="viridis")

# Improve readability
barplot.set_xticklabels(barplot.get_xticklabels(), rotation=90, fontsize = 8)  # Rotate the x-axis labels for better visibility
barplot.set_xlabel("Sample ID", fontsize=18)  # Rename x-axis label and set font size
barplot.set_ylabel("Reads in TSS", fontsize=18)  # Rename y-axis label and set font size
barplot.set_title("Proportion of Reads in TSS Across Samples", fontsize=20)  # Add title and set font size

plt.tight_layout()  # Adjust layout to make room for label rotation

plt.savefig("ATAC_seq_FRiTSS.png", dpi = 600)
plt.savefig("ATAC_seq_FRiTSS.svg", dpi = 600)
```
