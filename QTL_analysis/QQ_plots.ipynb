```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

base_dir = "http://bartzabel.ls.manchester.ac.uk/orozcolab/SNP2Mechanism/"
plt.rcParams['svg.fonttype'] = 'none'
```

## generate qqplot

```python
def generate(file, name):
    nominal_vals = pd.read_csv(file, sep = " ", header = None)
    p_values = nominal_vals.sample(10000)[11]
    expected_pvalues = np.linspace(1/(len(p_values)+1), 1, len(p_values))
    observed_log_pvalues = -np.log10(np.sort(p_values))
    expected_log_pvalues = -np.log10(expected_pvalues)

    # Create the QQ plot
    plt.figure(figsize=(10, 10))
    plt.scatter(expected_log_pvalues, observed_log_pvalues, c='blue', edgecolor='k', alpha=0.5)
    plt.plot([0, max(expected_log_pvalues)], [0, max(expected_log_pvalues)], color='red', linestyle='--')
    plt.xlabel('Expected -log10(p-value)')
    plt.ylabel('Observed -log10(p-value)')
    plt.title(f'QQ Plot of -log10(p-values) for {name}')
    plt.grid(True)
    plt.savefig(f"output_figs/QQ_plots/{name}_QTL_qqplot.svg")
```

```python
generate(f"{base_dir}/QTLs/RNA/RNA_nominal_CD8_no_min_chr1.txt", "RNA_CD8")
generate(f"{base_dir}/QTLs/RNA/RNA_nominal_CD4_no_min_chr1.txt", "RNA_CD4")

generate(f"{base_dir}/QTLs/HiC/ins_nominal_CD8_no_min_nominal_chr1.txt", "HiCins_CD8")
generate(f"{base_dir}/QTLs/HiC/ins_nominal_CD4_no_min_nominal_chr1.txt", "HiCins_CD4")

generate(f"{base_dir}/QTLs/HiC/loop_nominal_CD8_no_min_nominal_chr1.txt", "HiCloop_CD8")
generate(f"{base_dir}/QTLs/HiC/loop_nominal_CD4_no_min_nominal_chr1.txt", "HiCloop_CD4")

generate(f"{base_dir}/QTLs/ATAC/nominal_CD8_no_min_nominal_chr1.txt", "ATAC_CD8")
generate(f"{base_dir}/QTLs/ATAC/nominal_CD4_no_min_nominal_chr1.txt", "ATAC_CD4")
```
